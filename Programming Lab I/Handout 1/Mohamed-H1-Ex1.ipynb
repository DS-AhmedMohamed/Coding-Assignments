```python
import math

def estimate_pi():
    
    total_sum = 0
    k = 0
    summation_term = 1
    while summation_term >= 10**(-15):
        summation_term = (math.factorial(4*k)*(1103+26390*k))/(math.factorial(k)**4*396**(4*k))
        total_sum += summation_term
        k+=1

    pi_inverted = (2*math.sqrt(2))/9801*total_sum

    return 1 / pi_inverted

estimate_pi()
```

```python
estimate_pi() - math.pi
```

```python
def leibniz_pi():
    n = 0
    pi_inverted = 0
    counter = 0
    while round(pi_inverted*4,2) != 3.14:
        counter += 1
        pi_inverted += (-1)**n/(2*n+1)
        n += 1

    return print(f'estimated pi = {pi_inverted*4} and number of terms = {counter}')

leibniz_pi()
```

```python
import random

def monte_carlo(n_points):
    inside = 0
    outside = 0

    for i in range(n_points):
        p_x = random.uniform(-1.0,1.0)
        p_y = random.uniform(-1.0,1.0)

        if p_x**2 + p_y**2 <= 1:
            inside += 1
        else:
            outside += 1

    return 4*(inside/n_points)

numbers = [10, 100, 1000, 10000, 100000, 1000000]
for n in numbers:
    print(f'Pi = {monte_carlo(n)} for {n} points')
```
