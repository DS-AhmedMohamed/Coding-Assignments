(a) (2 pts) Find all happy numbers between 1 and 100 by writing a function is_happy(n)
that checks whether a number is happy or unhappy. It should return true if the the
number is happy and false otherwise.

```python
def IsHappy_while(n):

    while n not in [1,4]:
        print(f"n = {n}")
        n_str = str(n)
        sum = 0
        for i in n_str:
            sum += int(i) ** 2

        n = sum

    if n == 1 :
        return True
    else:
        return False
```

```python
IsHappy_while(500)
```

(b) (1 pt) Solve the problem (a) in two different ways using either while-loops or recursion.

```python
def IsHappy_recursion(n):

    print(f"n = {n}")
    n_str = str(n)
    sum = 0
    for i in n_str:
        sum += int(i) ** 2

    if sum == 1 :
        return True
    elif sum == 4: 
        return False
    else:
        IsHappy_recursion(sum)
```

```python
IsHappy_recursion(500)
```

(c) (2 pts) Modify (either the iterative or the recursive version of is_happy) to take an
additional argument p (i.e. def how_happy(n,p): and instead squaring digits raise
each digit to the power of p and then take the sum of the powers. Now, the following
different outcomes can be considered

```python
def how_happy(n,p):
    previous_numbers = []
    while n not in previous_numbers:
        previous_numbers.append(n)
        # print(f"n = {n}")
        n_str = str(n)
        sum = 0
        for i in n_str:
            sum += int(i) ** p
 
        if n>1 and n == sum:
            return "almost happy"
        else:
            n = sum
            

    if n == 1 :
        return "happy"
    else:
        return "unhappy"
```

```python
how_happy(153,3)
```

```python
how_happy(67888,3)
```

```python
happy = 0
almost = 0
unhappy = 0

for i in range(1,1001):
    result = how_happy(i,3)
    if result == "happy":
        happy += 1
    elif result == "almost happy":
        almost += 1
    else:
        unhappy += 1

print(f"Happy = {happy}, almost = {almost}, unhappy = {unhappy}")
```
