(1 pt) Write a function called has_duplicates that takes a list and returns True if
there is any element that appears more than once. It should not modify the original
list.

```python
def has_duplicates(l):
    '''return False if there is no duplicates and True if 2 or more elements has the same value
        parameters : 
        l : a collection - list or set
        return: boolean
    '''
    prev = list()

    for x in l:
        if x in prev:
            return True
        
        prev.append(x)

    return False
```

```python
def Three_non_unique(l):
    '''return True if more than 3 elements have non unique numbers in a list, Else, False'''
    prev = list()
    count = 0
    for x in l:
        if x in prev:
            count += 1
        
        prev.append(x)

    return False if count < 2 else True
```

```python
def BirthdayParadox(n, non_unique=2):
    '''conducts n number of birthday trials for 27 students and return the probability of non-unique values
    occurence in trials
    
    Parameters:
    n: integer'''
    assert non_unique in [2,3], "non_unique argument must be 2 or 3"
    from random import randint

    trials_results = []
    
    for t in range(n):
        rand_birthdays = [randint(1, 366) for x in range(27)]
        trials_results.append(Three_non_unique(rand_birthdays) if non_unique==3 else has_duplicates(rand_birthdays))

    return trials_results.count(True)/n
```

```python
def estimated_P(n,m):
    import math
    return 1 - math.exp(-(n**2/(2*m)))

def exact_P(n,m):
    pm_n = 1
    for i in range(1,n):
        pm_n = pm_n * ((m-i)/m)
    
    return 1 - pm_n
```
